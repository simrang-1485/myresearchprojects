```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#nodes data
nodes_data = pd.read_csv('dataset/a280_data.csv',header = None)
nodes_data
```

```python
#distance calculation 
def distance_cal(df):
    dist = []
    for i in range (0,len(df)):
        distances = []
        for j in range (0,len(df)):
            if i == j :
                distances.append(0.0)
            else :
                val = math.sqrt(math.pow(df.loc[i][1] - df.loc[j][1], 2) + math.pow(df.loc[i][2] - df.loc[j][2], 2));
                distances.append(val)
        dist.append(distances)
    return dist              
```

```python
df_data = distance_cal(nodes_data) #distance between all nodes
df_data
```

```python
dist_data = pd.DataFrame(df_data)
dist_data
```
